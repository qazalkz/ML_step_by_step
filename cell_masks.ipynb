```python
import cv2
import glob
from google.colab.patches import cv2_imshow
import numpy as np

path = "/content/drive/MyDrive/machine learning/cell dataset/masks/*.png"

def load_images_from_folder(path):
    path = glob.glob(path)
    masks = []
    for file in path:
        img = cv2.imread(file, cv2.IMREAD_UNCHANGED)
        masks.append(img)
        # cv2_imshow(img)
    return masks    
    
masks = load_images_from_folder(path)  
print(masks[1])
cv2_imshow(masks[50])
```

```python
print(masks[0].shape)
```

```python
w = masks[0].shape[0]
h = masks[0].shape[1]
unified_masks = np.zeros((w, h))
for mask in masks:
    unified_masks += mask
cv2_imshow(unified_masks)
```

```python
path_2 = "/content/drive/MyDrive/machine learning/cell dataset/images"

def show_image(path):
    path = glob.glob(path + "/*.png")
    
    for file in path:
        img = cv2.imread(file, cv2.IMREAD_UNCHANGED)
        cv2_imshow(img)


show_image(path_2)      
```
